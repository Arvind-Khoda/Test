```python
#Importing Required Libraries
import pandas as pd
import statistics as ss
import numpy as np
```

```python
#Importing Dataset
data=pd.read_csv("/content/country_wise_latest.csv")
data.columns
```

```python
#Droping Categorical Features
data.drop(columns=['Country/Region','WHO Region'],inplace=True)
```

```python
for col in data.columns:
     print(f"Measures of Central Tendency for {col} :")
     print("")
     print(f"Mean of {col} is {ss.mean(data[col])}")
     print(f"Harmonic mean of {col} is {ss.harmonic_mean(data[col])}")
     print(f"Mode of {col} is {ss.mode(data[col])}")
     print(f"Median of {col} is {ss.median(data[col])}")
     print("")
     print("")
     print(f"Measures of Dispersion for {col} :")
     print("")
     print(f"Variance of {col} is {ss.variance(data[col])}")
     print(f"Standard Deviation of {col} is {data[col].std}")
     print(f"Minimum Value of {col} is {np.percentile(data[col],0)}")
     print(f"25 Percentile of {col} is {np.percentile(data[col],25)}")
     print(f"50 Percentile of {col} is {np.percentile(data[col],50)}")
     print(f"75 Percentile of {col} is {np.percentile(data[col],75)}")
     print(f"Maximum Value of {col} is {np.percentile(data[col],100)}")
     print("")
     print("")
```
